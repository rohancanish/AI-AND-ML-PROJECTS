```python
import yfinance as yf
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt 
```

```python
stock = yf.download('AAPL', period='30d', interval='1d')

data = stock[['Close']]
data.reset_index(inplace=True)
```

```python
data.loc[:, 'Day'] = list(range(len(data)))

X = data[['Day']]
y = data['Close']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
next_day = np.array([[len(data)]])

predicted_price = model.predict(next_day)

print(f"Predicted Price for next day : $ {predicted_price[0].item():.2f}")
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.plot(data['Day'],y,label='Actual Price')
plt.plot(data['Day'],model.predict(X),label='Linear Fit')
plt.scatter(len(data),predicted_price,color='red',label='Predicted Next day')
plt.title("Stock Price Prediction")
plt.xlabel('Day')
plt.ylabel('Close Price')
plt.grid()
plt.legend()
plt.show()
```
